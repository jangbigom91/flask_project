```python
import plotly.graph_objects as go

df = pd.read_csv('PM10_seoul.csv')

fig = go.Figure(go.Scatter(x = df['Date'], y = df['Seoul'],
                          name='PM10'))

fig.update_layout(title='PM10 of Seoul',
                 plot_bgcolor='rgb(230, 230, 230)',
                 showlegend=True)

fig.show()
```
