```python
from dateutil.parser import parse
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('raw_dataset_plus.csv', parse_dates=['Date'], index_col='Date')
df.head()
```

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Tianjin', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    
plot_df(df, x=df.index, y=df.Tianjin, title='PM10 of Tianjin from 2014 to 2021.')
```

```python
df.reset_index(inplace=True)

df['year'] = [d.year for d in df.Date]
df['month'] = [d.strftime('%b') for d in df.Date]
```

```python
fig, axes = plt.subplots(1,2, figsize=(20, 7), dpi=100)
sns.boxplot(x='year', y='Seoul', data=df, ax=axes[0])
sns.boxplot(x='month', y='Seoul', data=df, ax=axes[1])

axes[0].set_title('년도별 박스도표', fontsize=18)
axes[1].set_title('월별 박스도표', fontsize=18)

plt.show()
```
